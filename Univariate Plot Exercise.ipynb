### Preliminary Steps:

```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loading in cleaned titanic data
df=pd.read_csv('/Users/oliver/Desktop/datasets/data-cleaning-core.csv')
display(df.info(),df.head())
```

### Univariate Graphs:

##### 1) Plot the counts for the number of passengers who Survived vs. those that died.


```python
# getting counts to plot
died=df['Survived'].loc[df['Survived']==0].count()
survived=df['Survived'].loc[df['Survived']==1].count()
# labels
bar_labels=['Survived','Died']
# plot
fig,ax=plt.subplots()
ax.bar(bar_labels,[died,survived])
ax.set(ylabel='Count',title='Passengers Survived vs Died');
```

##### 2) Plot a histogram of the distribution of Age.


```python
fig,ax=plt.subplots()
ax.hist(df.Age,9)
ax.set(xlabel='Age',ylabel='Count',title='Distribution of Age');
```

##### 3) Plot a boxplot for the Fare column.



```python
fig,ax=plt.subplots()
ax.boxplot(df['Fare'])
ax.set(xlabel='Fare',ylabel='Distribution',title='Box Plot of Fare Distribution');
```
