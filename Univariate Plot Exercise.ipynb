### Preliminary Steps:

```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loading in cleaned titanic data
df=pd.read_csv('/Users/oliver/Desktop/datasets/data-cleaning-core.csv')
display(df.info(),df.head())
```

### Univariate Graphs:

##### 1) Plot the counts for the number of passengers who Survived vs. those that died.


```python
# survival count
survived=df['Survived'][df['Survived']==1].count()
# death count
died=df['Survived'][df['Survived']==0].count()
# labels for both
bar_labels=['survived','died']
# plot
fig,ax=plt.subplots()
ax.bar(bar_labels,[survived,died])
# set title and y-label
ax.set(ylabel='Count',title='Passenger Count: Survived vs. Died');
```

##### 2) Plot a histogram of the distribution of Age.


```python
fig,ax=plt.subplots()
ax.hist(df.Age,9)
ax.set(xlabel='Age',ylabel='Count',title='Distribution of Age');
```

##### 3) Plot a boxplot for the Fare column.



```python
fig,ax=plt.subplots()
ax.boxplot(df['Fare'])
ax.set(xlabel='Fare',ylabel='Distribution',title='Box Plot of Fare Distribution');
```
